```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from datetime import date
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
df = spark.read.option('header', 'true').csv('videos-stats.csv')
```

```python
df.show(8)
```

```python
df.printSchema()
```

```python
df = spark.read.option('header', 'true').option('InferSchema', 'true').csv('videos-stats.csv')
```

```python
df.printSchema()
```

```python
df.write.option('header', 'true').save('drive/MyDrive/Colab Notebooks/videos-parquet')
```

```python
df = spark.read.option('header', 'true').parquet('drive/MyDrive/Colab Notebooks/videos-parquet')
```

```python
df.write.option('header', 'true').saveAsTable('tb_videos')
```

```python
spark.catalog.listTables()
```

```python
tab_df = spark.sql('select * from tb_videos')
print(tab_df.show(8))
```
